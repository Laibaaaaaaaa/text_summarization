```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

```

```python
model = T5ForConditionalGeneration.from_pretrained('t5-small')  # or 'facebook/bart-large-cnn' for BART
tokenizer = T5Tokenizer.from_pretrained('t5-small')
```

```python
def summarize(text):
    # Preprocess input text for the model
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)

    # Generate summary (set maximum and minimum length for output)
    summary_ids = model.generate(inputs, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)

    # Decode and return the summary
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
text = """
Here, you can enter a long piece of text or paste an article for testing. The function will take this input and provide a concise summary.
"""
summary = summarize(text)
print("Summary:", summary)
```

```python
# Run this cell to prompt for user input
user_text = input("Enter text to summarize: ")
summary = summarize(user_text)
print("Summary:", summary)
```
